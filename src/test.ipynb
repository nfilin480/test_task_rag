Создать базу знаний википедии (100к примеров будут обрабатываться примерно час)

```python
import pandas as pd
from vectorestore import KnowledgeDatabase
from embedder import E5Embeddings

db = KnowledgeDatabase(
    embeddings=E5Embeddings(),
    chunk_size=512,  # размер чанка в токенах
    chunk_overlap=20  # перекрытие между чанками в токенах
)
```

```python
db.load_from_csv("./wikipedia_100k.csv", "text")
```

Тестирование RAG системы


```python
from main import RAGSystem
```

```python
print("Инициализация RAG системы...")
rag_system = RAGSystem()
print("Система готова к работе!")
print("Введите 'выход' для завершения работы")

while True:
    query = input("\nВаш вопрос: ")
    
    if query.lower() in ['выход', 'exit', 'quit']:
        break
        
    try:
        result = rag_system.process_query(query)
        if "error" in result:
            print(f"\nПроизошла ошибка: {result['error']}")
        else:
            print(f"\nТип запроса: {result['query_type']}")
            print(f"Ответ: {result['answer']}")
    except Exception as e:
        print(f"Произошла ошибка: {str(e)}")
```
