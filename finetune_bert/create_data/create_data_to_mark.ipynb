```python
from datasets import load_dataset
import pandas as pd

dataset = load_dataset("yahma/alpaca-cleaned")

df = pd.DataFrame(dataset['train'])

df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Создаем новый столбец с общей длиной input и instruction
df['total_length'] = df['input'].fillna('').str.len() + df['instruction'].str.len()

# Фильтруем строки, где общая длина меньше 2000 символов
df_filtered = df[df['total_length'] < 1300].reset_index(drop=True)

df_filtered[:3000].to_csv('./datasets/alpaca.csv', index=False)
```
